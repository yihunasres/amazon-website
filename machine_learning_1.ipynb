```python
import pandas as pd
# from sklearn import logesticRegration
from sklearn.model_selection import train_test_split
df = pd.read_csv('/content/drive/MyDrive/machine_learning/Student_performance_data _.csv')

x=df[['StudentID','Age','Gender',	'Ethnicity','ParentalEducation','StudyTimeWeekly','Absences','Tutoring','ParentalSupport','Extracurricular','Sports','Music','Volunteering','GPA']]
y=df['GradeClass']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
len(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
linear_model=LogisticRegression()
linear_model.fit(x_train,y_train)
```

```python
linear_model.predict(x_test)
```

```python
y_test
```

```python
linear_model.score(x_test,y_test)
```
